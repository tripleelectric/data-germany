```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('default')
plt.rcParams['figure.figsize'] = (10, 6)
plt.rcParams['font.size'] = 12
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd',
          '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf']
```

```python
from numpy import float16


df = pd.read_csv('data/raw/12612-0019_de_flat.csv', delimiter=';')
df = df.loc[:, ['time', 'value']]
# df['value'] = df['value'].apply(lambda x: x.replace(',', '.'))
# df['value'].astype(float16)
df['value'] = pd.to_numeric(df['value'].str.replace(',', '.'))
df = df.sort_values(by='time')
df.head()
```

```python
df.plot(
    x='time',
    y='value',
    kind='line',  # Change 'scatter' to 'line'
    marker='o',   # Still a good idea to mark the actual data points
    figsize=(10,6),
    title='Rate over Years',
    xlabel='Year',
    ylabel='Birth Rate per 1000 citizens'
)

plt.grid(True)
plt.tight_layout()
plt.show()
```

The baby boomer peak is visible, followed by the decline impartly caused by the invention of new contraceptives.
